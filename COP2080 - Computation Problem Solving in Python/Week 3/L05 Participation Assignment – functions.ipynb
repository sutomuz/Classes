**5.** Create a docstring for scoreToNumber function in the code below. Use a format similar to this to summarize the function, describe the parameters, data types, and returns. Show your code and output. 

```python
def scoreToNumber(score):
    """
    Converts a text-based score into a numeric value.

    Parameters:
        score (str): A string representing a score.
                     Expected values are:
                     Great, Ok, or Poor (case-insensitive)

    Returns:
        result (int): Numeric score based on input:
                      Great = 10
                      Ok    = 5
                      Poor  = 3
    """
    score = str.upper(score)
    result = 0
    first = score[0]
    if first == "G" :
        result = 10
    elif first == "O" :
        result = 5
    elif first == "P" :
        result = 3
    return result

def main():
    score1 = input('Enter score 1 as Great, Ok or Poor: ')
    score2 = input('Enter score 2 as Great, Ok or Poor: ')
    score3 = input('Enter score 3 as Great, Ok or Poor: ')
    x1 = scoreToNumber(score1)
    x2 = scoreToNumber(score2)
    x3 = scoreToNumber(score3)
    xmax = max(x1, x2, x3)
    some_avg = (x1 + x2 + x3) / 3
    print(f'The maximum score was {xmax}')
    print(f'The avg score on 1-10 scale would have been {round(some_avg, 2)}')

main()
#
# Double click to copy code
```
