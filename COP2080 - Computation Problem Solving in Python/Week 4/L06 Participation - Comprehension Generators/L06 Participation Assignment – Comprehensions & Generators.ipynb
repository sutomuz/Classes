**1.** Modify the following code to use both functions (strip_string and uppercase_decorator) as decorators. Show your modified code and output:

```python
# Modified Code
# strip off dashes
def strip_string(b_function):
    def wrapper():
        func = b_function()
        strip_string = func.strip('-')
        return strip_string
    return wrapper

# create upper case version of clli code
def uppercase_decorator(some_function):
    def a_wrapper():
        func = some_function()
        make_uppercase = func.upper()
        return make_uppercase
    return a_wrapper

@uppercase_decorator
@strip_string
def clli_code():
    print("The Florida router clli code is", end=" ")
    return '--jzrtpaflxacg19----'

print(clli_code())
```

**6.** Correct the code implement a generator comprehension for ‘b’ that produces the same list of 1 million values as does the list comprehension. Generator expressions use () instead of [] braces.
Note the difference in memory using a list vs a generator. Show your code and output:


```python
# listcomp_vs_genexp.py
# list comprehension vs generator expression

import sys
a = [x for x in range(1000000)] # list comp
b = (x for x in range(1000000)) #generator expr
print('list comp byte size is', sys.getsizeof(a))
print('generator expression byte size is', sys.getsizeof(b))
```
